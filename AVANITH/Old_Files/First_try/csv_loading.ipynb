```python
import pandas as pd

# Load original CSV
input_path = r"C:\Users\kanam\OneDrive\Desktop\Visualization_Literacy\Visualization_Literacy\Avanith\Final_chartgemma__Determine_Range.csv"
df = pd.read_csv(input_path)

# Extract price range from the *answer sentence only*
df['Prediction'] = df['Prediction'].str.extract(
    r'(?:was|between)\s*\$?(\d+\.\d+)[\s\-andto]*\$?(\d+\.\d+)', expand=True
).agg(lambda x: f"${x[0]} - ${x[1]}", axis=1)

# Save cleaned CSV
output_path = r"C:\Users\kanam\OneDrive\Desktop\Visualization_Literacy\Visualization_Literacy\Avanith\cleaned_csv.csv"
df.to_csv(output_path, index=False)

# Preview
from IPython.display import display
pd.set_option('display.max_colwidth', None)
display(df[['Prediction']].head())
```
