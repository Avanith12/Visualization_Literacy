# 6.Visualize line dataset for each question
 

```python
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import os

# Base path where images are stored
base_path = "/Users/ramihuunguyen/Downloads/osfstorage-archive/generated_plots"

# Make sure output folder exists
output_dir = "color_chart_outputs"
os.makedirs(output_dir, exist_ok=True)

# Filter 16 unique questions
target_questions = df['question'].dropna().unique()[:16]

# Loop over each question
for q_idx, target_question in enumerate(target_questions):
    matching_rows = df[df['question'] == target_question].copy()

    if matching_rows.empty:
        continue

    # Identify color columns
    color_cols = [col for col in matching_rows.columns if col.startswith("color_")]

    # Create a unique color combination key
    matching_rows['color_combo'] = matching_rows[color_cols].apply(lambda row: tuple(row), axis=1)
    unique_color_rows = matching_rows.drop_duplicates(subset='color_combo')

    # Get the answer for this question
    answer = matching_rows['answer'].iloc[0] if 'answer' in matching_rows.columns else "N/A"

    # Plot setup
    fig, axes = plt.subplots(2, 5, figsize=(22, 10))
    axes = axes.flatten()

    for i, (_, row) in enumerate(unique_color_rows.iterrows()):
        if i >= len(axes):
            break

        img_path = f"{base_path}/{row['filename']}"

        # Get readable color names
        colors_used = [col.replace("color_", "") for col in color_cols if row[col]]
        title_text = ", ".join(colors_used) if colors_used else "No Color"

        try:
            img = Image.open(img_path)
            axes[i].imshow(img)
            axes[i].set_title(title_text, fontsize=10)
            axes[i].axis('off')
        except FileNotFoundError:
            axes[i].text(0.5, 0.5, "Image not found", ha='center', va='center', fontsize=8)
            axes[i].set_title("Missing", fontsize=10)
            axes[i].axis('off')

    # Set main title
    fig.suptitle(f"{target_question}\nAnswer: {answer}", fontsize=15, y=1.02)
    plt.tight_layout()

    # Save chart
    filename_safe = f"q{q_idx+1:02d}_line_chart_color_usage.png"
    plt.savefig(os.path.join(output_dir, filename_safe), dpi=300, bbox_inches='tight')
    plt.close()

print("✅ All 16 charts saved!")
```
