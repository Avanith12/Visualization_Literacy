### 2: Run one example of line charts

```python
from Gemini.gemini_1_5_flash import GeminiVisionRunner

# Test parameters
image_file = "test image from make comparision.png"
prompt_text = "About how much did the price of a barrel of oil fall from April to September in 2015? Options: $4, $15, $17, $45, Omit. Answer only. No explanation."

# Initialize and run
runner = GeminiVisionRunner(
    image_path=image_file,
    prompt=prompt_text,
    runs=3
)

runner.run()
```

### 3. Results for 3 times run

```python
import pandas as pd

# Read lines from the output file
with open("gemini_output.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()

# Parse outputs
runs = []
answers = []

for i in range(0, len(lines), 3):
    run_label = lines[i].strip().replace("🔁 ", "")
    answer = lines[i+1].strip()
    runs.append(run_label)
    answers.append(answer)

# Create DataFrame
df = pd.DataFrame({
    "Run": runs,
    "Gemini Answer": answers
})

# Display as table
df.style.set_caption("📊 Gemini Answers Across 3 Runs")
```
